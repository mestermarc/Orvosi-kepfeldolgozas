```python
from skimage import filters

tumor = []

tumor.append(one)
tumor.append(two)
tumor.append(three)
tumor.append(four)

for slices in tumor:
    plt.imshow(slices)
    plt.show()
```

```python
edges = []

for slices in tumor:
    edge = filters.roberts(slices)
    edge[edge > 0] = 1
    int_array = edge.astype(int)
    edges.append(int_array)
    plt.imshow(int_array)
    plt.show()
print(len(edges))
```

```python
plt.imshow(edges[1])
```

```python
pontfelho = []
slicecounter = 0
for edge in edges:
    print("Len:", len(edge))
    #print(edge)
    middlex = len(edge)
    for row in range(0,len(edge)):
        for column in range(0,len(edges[slicecounter])):
            #print(row, column, slicecounter)
            if edge[row][column]>0:
                pontfelho.append(([row],[column],[slicecounter]))
    slicecounter+=1
print(len(pontfelho))
print(pontfelho[12])
```

```python
xp=[]
yp=[]
zp=[]

for pont in pontfelho:
    xp.append(pont[0])
    yp.append(pont[1])
    zp.append(pont[2])


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xp, yp, zp, zdir='z', s=20, c='b',rasterized=True)
ax.set_xlabel('x')

plt.show()
```
